```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('agg').getOrCreate()
```

```python
one = spark.read.csv('file:/Workspace/Users/vaishnavikumbhakarna467@gmail.com/practice_pyspark/test4.csv',header=True,inferSchema=True)
display(one)
```

```python
two = one.groupBy('Name').avg('Salary').show()
two = one.groupBy('Name').avg().show()
```

```python
two = one.groupBy('Name').min('Salary').show()
```

```python
two = one.groupBy('Name').max('Salary').show()
```

```python
two = one.groupBy('Departments').avg('Salary').show()
two = one.groupBy('Departments').avg().show()
```

```python
two = one.groupBy('Departments').sum('Salary').show()
two = one.groupBy('Departments').mean('Salary').show()
two = one.groupBy('Departments').count().show()
```

```python
three = one.agg({'Salary':'sum'})
display(three)
```
