Drop function has 3 parameters 
How
Thresh 
Subset

```python
df_new=df_pyspark.na.drop(how="any",thresh=2)
display(df_new)
```

```python
df_one = df_pyspark.na.drop(how="any",subset=["Age"])
display(df_one)
```

```python
df_two = df_pyspark.na.drop(how="all")
display(df_two)
```

In the **fill** function the first parameter is the value to be filled and the second parameter should be a list of columns (or a single column name in some versions), but the value you're filling must also match the column's data type.

```python
df_three = df_pyspark.na.fill("Value Missing")
display(df_three)
```

```python
df_four = df_pyspark.na.fill(11111,['Age','Experience'])
display(df_four)
```

## Imputer function


```python
from pyspark.ml.feature import Imputer 
```

```python
imputer = Imputer(
    inputCols = ['Age','Experience','Salary'],
    outputCols= ["{}_imputed".format(col) for col in ['Age','Experience','Salary']]
).setStrategy("median")
```

```python
imputer.fit(df_pyspark).transform(df_pyspark).show()
# display(df_pyspark)
```
